## Glove Word Embeddings

```python
glove_file = project_path + "glove.6B.zip"
```

```python
#Extract Glove embedding zip file
from zipfile import ZipFile
with ZipFile(glove_file, 'r') as z:
  z.extractall()
```

# Word Embeddings using Embedding file as given below.

```python
EMBEDDING_FILE = './glove.6B.200d.txt'

embeddings = {}
for o in open(EMBEDDING_FILE):
    word = o.split(" ")[0]
    # print(word)
    embd = o.split(" ")[1:]
    embd = np.asarray(embd, dtype='float32')
    # print(embd)
    embeddings[word] = embd

```

```python
token.word_index.items()
```

```python
len(token.word_index.items())
```

# Creating a weight matrix for words in training docs

```python
embedding_matrix = np.zeros((num_words+1, 200))
```

```python
#embedding_matrix[29656]
```

```python
for word, i in token.word_index.items():
    embedding_vector = embeddings.get(word)
    if embedding_vector is not None:
        #print (i)
        embedding_matrix[i] = embedding_vector

len(embeddings.values())
```

## Creating and Compiling the Model

```python
model = Sequential()
model.add(Embedding(num_words+1, embedding_size, weights = [embedding_matrix]))
model.add(Bidirectional(LSTM(128, return_sequences = True)))
model.add(Dropout(0.1))
model.add(Dense(250, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
batch_size = 100
epochs = 5
```

```python
history = model.fit(X,y, batch_size=batch_size, epochs=epochs, validation_split=0.2, verbose=1) # Train
```

### *The validation accuracy of the model is*  **85.91%**
